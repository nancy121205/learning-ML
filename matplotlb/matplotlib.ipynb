## Matplotlib: Pyplot API vs Object-Oriented (OO) API

---

###  **1. Pyplot API (`matplotlib.pyplot`)  State-based**

####  Description:

* Mimics MATLAB.
* Uses **implicit state**: tracks the current figure and axes.
* You dont manually manage `Figure` or `Axes` objects.
* Functions like `plt.plot()`, `plt.title()` apply to the current plot.

####  Pros:

* Simple and intuitive.
* Great for quick plots and data exploration.
* Ideal for Jupyter notebooks or small scripts.

####  Cons:

* Hard to manage complex plots (e.g. subplots).
* Not modular  hard to scale or reuse.

####  Example:

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Pyplot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

---

###  **2. Object-Oriented (OO) API  Explicit**

####  Description:

* You create and control `Figure` and `Axes` objects directly.
* Plotting is done using methods like `ax.plot()`, `ax.set_title()`, etc.
* **More Pythonic** and flexible.

####  Pros:

* Better control over layout, styles, and multiple plots.
* Cleaner and more maintainable code.
* Ideal for professional applications, subplots, GUIs, dashboards.

####  Cons:

* Slightly more verbose.
* Takes more effort to learn at the beginning.

####  Example:

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("OO Example")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.grid(True)
plt.show()
```

---

###  **Comparison Table**

| Feature              | Pyplot API                      | Object-Oriented API                      |
| -------------------- | ------------------------------- | ---------------------------------------- |
| **Control Level**    | Implicit (state-based)          | Explicit (you control everything)        |
| **Ease of Use**      | Easier for beginners            | More powerful, modular                   |
| **Subplots Support** | Limited / repetitive            | Clean and flexible                       |
| **Scalability**      | Not ideal for complex projects  | Excellent for large-scale applications   |
| **Best Use Case**    | Quick testing, small scripts    | Dashboards, apps, research projects      |
| **Axes Handling**    | Handled internally by `plt`     | You create and manage axes               |
| **Looks like**       | `plt.plot(...); plt.title(...)` | `fig, ax = plt.subplots(); ax.plot(...)` |

---

## two options for subplots

````python
#1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, 
                               ncols = 2, 
                               figsize = (9,4))
ax1.plot(x,x**2)
ax2.bar(x,x)
ax3.scatter(x, x**2)
ax4.hist(np.random.randn(1000));
````

````python
#2
fig, ax = plt.subplots(nrows = 2, 
                               ncols = 2, 
                               figsize = (9,4))
ax[0,0].plot(x,x**2)
ax[0,1].bar(x,x)
ax[1,0].scatter(x, x**2)
ax[1,1].hist(np.random.randn(1000));
````

````python
over60 = hd[hd.age > 60]
over60
````

````python
# pyplot method
over60.plot.scatter(x = 'age', y = 'chol', c = 'target')
````

````python
# OO method mixed with pyplot method
# Create a Figure and Axes instance
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data from the DataFrame to the ax object
over60.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target", 
             ax=ax); # set the target Axes
````

````python
fig, ax = plt.subplots(figsize=(10, 6))

over60.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target", 
             ax=ax);

ax.set_xlim([60, 70]);         # set axesses of x and y
````

````python
fig, ax = plt.subplots(figsize = (10,6))

# plot data
scatterplot = ax.scatter(x = over60.age,
                         y = over60.chol,
                         c = over60.target)

ax.set(xlabel = "Age",
       ylabel = "Cholestrol",
       title = "Heart Disease and Cholestrol Level")

ax.legend(*scatterplot.legend_elements(), title = "Target");
````

````python
fig, ax = plt.subplots(figsize = (10,6))

# plot data
scatterplot = ax.scatter(x = over60.age,
                         y = over60.chol,
                         c = over60.target)

# add labels
ax.set(xlabel = "Age",
       ylabel = "Cholestrol",
       title = "Heart Disease and Cholestrol Level")

# add legend
ax.legend(*scatterplot.legend_elements(), title = "Target");

# add horizontal line for mean
ax.axhline(over60.chol.mean(), linestyle = '--');
````

````python
# subplots of age, thalach, chol
fig, (ax0, ax1) = plt.subplots(nrows = 2, 
                               ncols = 1, 
                               figsize = (10, 6),
                               sharex = True)

plot0 = ax0.scatter(x = over60.age, y = over60.chol, c = over60.target)
plot1 = ax1.hist(x = over60.age, color = "purple")

ax0.set(ylabel = "Cholestrol", title = "Age and Cholestrol")
ax1.set(xlabel = "Age", title = "Age Distribution");

fig.subplots_adjust(hspace=0.4);
fig.suptitle("Heart Disease visualiztion", fontweight = "bold", fontsize = 20);

ax0.axhline(over60.chol.mean(), linestyle = '--');
````
